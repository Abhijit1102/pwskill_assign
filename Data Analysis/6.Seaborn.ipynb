## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

`Seaborn` is a popular data visualization library in Python, which provides a wide range of plots for data analysis. Here are five plots that we can plot using Seaborn and their uses:

1. `Scatter plot` - `Scatter plot` is a plot of two variables showing their correlation. We can use a scatter plot to identify the relationship between two continuous variables.

2. `Histogram` - `Histogram` is a plot showing the distribution of a single variable. We can use a histogram to check the data's skewness, the spread of data, and the location of the data.

3. `Heatmap` - `Heatmap` is a plot showing the correlation matrix between variables. We can use a heatmap to visualize the correlation between multiple variables, making it easier to identify the patterns in the data.

4. `Boxplot` - `Boxplot` is a plot showing the distribution of a single variable across different categories. We can use a boxplot to identify the outliers, quartiles, and skewness in the data.

5. `Line plot` - `Line plot` is a plot showing the trend of a variable over time. We can use a line plot to identify the changes in the variable over time and compare the trends of multiple variables.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')

sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri)

plt.title('Signal over time for different events and regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.grid()

plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
## Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age distribution by passenger class')
plt.xlabel('Passenger class')
plt.ylabel('Age')

plt.subplot(1,2,2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare distribution by passenger class')
plt.xlabel('Passenger class')
plt.ylabel('Fare')

plt.suptitle('Box plots of age and fare by passenger class')
plt.tight_layout()

plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')

sns.histplot(data=diamonds, x='price', hue='cut', kde=True, multiple='stack')

plt.title('Price distribution by cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.grid()
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset('iris')

sns.pairplot(data=iris, hue='species')

plt.suptitle('Pair plot of iris dataset by species')

plt.grid()
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')

flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

sns.heatmap(data=flights_pivot, cmap='summer')

plt.title('Passenger counts by year and month')

plt.show()
```
