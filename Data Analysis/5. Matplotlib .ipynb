## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

`Matplotlib` is a Python data visualization library that allows users to create static, animated, and interactive visualizations in Python. It is widely used in data science, scientific computing, and engineering applications to create visualizations such as line plots, scatter plots, bar plots, histograms, and more.

`Matplotlib` is used because it provides a high degree of control over every aspect of a visualization, from the axis labels and tick marks to the line styles and colors. It is also highly customizable, allowing users to create professional-quality visualizations with ease.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. `Line Plot`: A line plot is used to visualize the trend of a continuous variable over time or other continuous variable. It can be created using the plot() function in Pyplot.

2. `Scatter Plot`: A scatter plot is used to visualize the relationship between two continuous variables. It can be created using the scatter() function in Pyplot.

3. `Bar Plot`: A bar plot is used to visualize the distribution of a categorical variable or the relationship between a categorical variable and a continuous variable. It can be created using the bar() or barh() function in Pyplot.

4. `Histogram`: A histogram is used to visualize the distribution of a continuous variable. It can be created using the hist() function in Pyplot.

5. `Box Plot`: A box plot is used to visualize the distribution of a continuous variable across different categories. It can be created using the boxplot() function in Pyplot.





## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
## import numpy as np
## np.random.seed(3)
## x = 3 + np.random.normal(0, 2, 50)
## y = 3 + np.random.normal(0, 2, len(x))
## Note: Also add title, xlabel, and ylabel to the plot.

A `scatter` plot is a type of plot in which the values of two variables are plotted along two axes, with each data point represented as a point on the plot. The position of each point on the plot is determined by the values of the two variables for that data point.

A `scatter` plot is useful for visualizing the relationship between two variables. If the variables are positively correlated, the points on the scatter plot will tend to form a pattern that slopes upward from left to right. If the variables are negatively correlated, the points on the scatter plot will tend to form a pattern that slopes downward from left to right. If there is no correlation between the variables, the points will be scattered randomly across the plot.

`Scatter` plots can also be used to identify outliers, which are data points that are significantly different from the rest of the data. Outliers are usually represented as points that are far away from the other points on the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,4))
plt.scatter(x,y,c="#FF2743", alpha=0.6)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.show()
```

The `subplot()` function in Matplotlib is used to create multiple plots in the same figure. It takes three arguments: the number of rows of plots, the number of columns of plots, and the plot number (starting from 1) that the current plot will occupy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the data
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plot the data
for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.bar(eval(f'x{i}'), eval(f'y{i}'), color="#FF9821", alpha=0.5)
    plt.title(f'Line {i}')
    plt.xlabel('X-Axis')
    plt.ylabel('Y-Axis')
    plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
## import numpy as np
## company = np.array(["Apple", "Microsoft", "Google", "AMD"])
## profit = np.array([3000, 8000, 1000, 10000])


A `bar plot` is a type of chart that displays data as a series of rectangular bars, where the height or length of each bar is proportional to the value it represents. Bar plots can be vertical (with the bars arranged from top to bottom) or horizontal (with the bars arranged from left to right).

`Bar plots` are commonly used to compare the values of different categories or groups of data. They are particularly useful when the data has a natural order or hierarchy, such as time periods or levels of a categorical variable. Bar plots are also useful for showing how data is distributed across different categories or groups, as well as for visualizing the differences or changes between these groups.

`Bar plots` can be created using various tools and libraries, such as Matplotlib, Seaborn, or Pandas in Python. They are widely used in fields such as business, finance, economics, marketing, and social sciences for data analysis and presentation.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(5,6))
plt.bar(company, profit, color="#FF2473", alpha=0.5)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company's Profit")
plt.grid()
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
## box1 = np.random.normal(100, 10, 200)
## box2 = np.random.normal(90, 20, 200)

A `box plot ` is a graphical representation of a dataset that displays the distribution of the data based on the five number summary: minimum, lower quartile, median, upper quartile, and maximum. The box in the plot represents the middle 50% of the data, with the median (the middle value of the dataset) represented by a line inside the box. The whiskers extend from the box to the minimum and maximum values, excluding any outliers, which are represented as individual points beyond the whiskers.



```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8,5))
plt.subplot(2, 2, 1)
plt.boxplot(box1)
plt.title('Boxplot 1')

plt.subplot(2, 2, 2)
plt.boxplot(box2)
plt.title('Boxplot 2')

plt.show()
```
