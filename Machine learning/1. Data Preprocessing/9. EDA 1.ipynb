The wine quality data set contains 12 independent variables that describe different chemical properties of wines and one dependent variable that indicates the quality of wine on a scale of 0-10. The key features of the wine quality data set are:

- 1. `Fixed acidity`: It is the amount of tartaric acid present in the wine. It is an important factor in determining the acidity of wine.

- 2. `Volatile acidity`: It is the amount of acetic acid present in the wine. High levels of volatile acidity can make wine taste unpleasant and vinegary.

- 3. `Citric acid`: It is the amount of citric acid present in the wine. Citric acid can add a refreshing tartness to the wine.

- 4. `Residual sugar`: It is the amount of sugar that remains in the wine after the fermentation process. It can contribute to the sweetness of wine.

- 5. `Chlorides`: It is the amount of salt present in the wine. High levels of chlorides can make wine taste salty.

- 6. `Free sulfur dioxide`: It is the amount of sulfur dioxide that is added to wine as a preservative. It can prevent the growth of bacteria and yeast.

- 7. `Total sulfur dioxide`: It is the total amount of sulfur dioxide present in the wine, including the free and bound forms.

- 8. `Density`: It is the mass per unit volume of the wine. It can indicate the alcohol content of the wine.

- 9. `pH`: It is a measure of the acidity or alkalinity of the wine. A lower pH indicates higher acidity.

- 10. `Sulphates`: It is the amount of sulfur compounds that are added to wine during the winemaking process. Sulphates can help to prevent spoilage and oxidation of wine.

- 11. `Alcohol`: It is the percentage of alcohol present in the wine. It can affect the flavor and aroma of wine.

- 12. `Quality`: It is the dependent variable that indicates the quality of wine on a scale of 0-10.

## Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

```python
## Checking Missing Value in the Datasets
df.isnull().sum()
```

Some common techniques for handling missing data and their advantages and disadvantages.

- 1. `Deletion`: In this technique, missing data is simply deleted from the dataset. There are two types of deletion methods: listwise deletion (deletes entire rows with missing data) and pairwise deletion (deletes only the missing values in each variable pair). The advantage of this technique is that it is simple and can be effective if the proportion of missing data is small. The disadvantage is that it can reduce the sample size and may result in biased estimates if data is not missing at random.

- 2. `Mean imputation`: In this technique, missing values are replaced with the mean value of the variable. The advantage is that it is simple and can preserve the sample size. However, it can lead to biased estimates if data is not missing at random and can also underestimate the true variance of the variable.

- 3. `Regression imputation`: In this technique, a regression model is used to predict the missing values based on other variables in the dataset. The advantage is that it can be more accurate than mean imputation and can preserve the sample size. However, it requires a large sample size and can be sensitive to outliers and non-linear relationships.

- 4. `Multiple imputation`: In this technique, multiple plausible values are imputed for each missing value, and the results are combined to provide estimates and standard errors that reflect the uncertainty due to missing data. The advantage is that it can provide more accurate estimates than single imputation methods and can incorporate the uncertainty due to missing data. However, it requires advanced statistical software and can be computationally intensive.

- 2. Next, we could use `inferential statistics` to test for significant differences between groups. For example, we could use a t-test to compare the mean math scores between male and female students, or a one-way ANOVA to test for differences in reading scores across different racial/ethnic groups.

```python
import scipy.stats as stats
```

```python
# Extract math scores for male and female students
male_math_scores = df.loc[df['gender'] == 'male', 'Total_score']
female_math_scores = df.loc[df['gender'] == 'female', 'Total_score']

# Perform a t-test for independent samples
t, p = stats.ttest_ind(male_math_scores, female_math_scores)
print("t = ", t)
print("p-value = ", p)
```

## Interpretation of 't-test'

The `t-value of` -4.1699 suggests that the difference in total scores between male and female students is statistically significant. The negative t-value indicates that the mean total score is higher for females compared to males.

The `p-value` of 3.3119e-05 is less than the commonly used significance level of 0.05, which indicates strong evidence against the null hypothesis (that there is no difference in total scores between males and females). In other words, we can reject the null hypothesis and conclude that there is a significant difference in total scores between male and female students.

```python
# Perform a one-way ANOVA
result = stats.f_oneway(df.loc[df['race_ethnicity'] == 'group A', 'reading_score'],
                        df.loc[df['race_ethnicity'] == 'group B', 'reading_score'],
                        df.loc[df['race_ethnicity'] == 'group C', 'reading_score'],
                        df.loc[df['race_ethnicity'] == 'group D', 'reading_score'],
                        df.loc[df['race_ethnicity'] == 'group E', 'reading_score'])
print("F-value = ", result.statistic)
print("p-value = ", result.pvalue)
```

## Interpretation of Anova
The `F-value` of 5.6217 suggests that there is a statistically significant difference in reading scores between at least one pair of the groups (i.e. different racial/ethnic groups).

The `p-value` of 0.0001780 is less than the commonly used significance level of 0.05, which indicates strong evidence against the null hypothesis (that there is no difference in reading scores across the different racial/ethnic groups). In other words, we can reject the null hypothesis and conclude that there is a significant difference in reading scores across different racial/ethnic groups.

- 3. We could also use `regression analysis` to model the relationship between various factors and exam performance. For instance, we could build a multiple linear regression model to predict writing scores based on variables such as parental education level, lunch status, and test preparation course completion.

```python
import statsmodels.api as sm

# Extract predictor variables and response variable
X = df[['parental_level_of_education', 'lunch', 'test_preparation_course']]
y = df['writing_score']

# Encode categorical variables as dummy variables
X = pd.get_dummies(X, drop_first=True)

# Add intercept term to predictor variables
X = sm.add_constant(X)

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Print summary of model results
print(model.summary())
```

## Interpretation of Regression 
This is a summary of a multiple linear regression model built to predict writing scores based on several variables. The model has an `R-squared` value of 0.228, which indicates that 22.8% of the variance in writing scores can be explained by the predictors included in the model.

The results show that `parental education level`, `lunch status`, and test preparation course completion are all significant predictors of writing scores. Specifically, having a bachelor's degree, a master's degree, or some high school education level of parental education is associated with higher writing scores, while having a high school education level is associated with lower writing scores. Additionally, students who receive standard lunches score higher on writing, while those who do not complete a test preparation course score lower.

The `F-statistic` is significant (p < 0.001), indicating that the overall model is a good fit for the data. The coefficients for each predictor variable represent the estimated change in the dependent variable for a one-unit change in that predictor variable, while holding all other predictor variables constant.

The Omnibus test for normality and the Jarque-Bera test for symmetry indicate that the residuals of the model are approximately normally distributed. The Durbin-Watson statistic (2.010) suggests that there is no significant autocorrelation in the residuals. The condition number (8.58) suggests that there is no significant multicollinearity among the predictor variables.

Overall, this model suggests that parental education level, lunch status, and test preparation course completion are important predictors of writing scores, and could be used to inform interventions aimed at improving student writing performance.

## Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?

```python
import pandas as pd
df = pd.read_csv("winequality-red.csv")
df.head()
```

```python
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
normal_features = []
non_normal_features = []


for column in df.columns:
    # Histogram
    plt.subplots(1,2,figsize=(10,4))
    plt.subplot(1,2,1)
    sns.histplot(df[column], kde=True)
    plt.title(column)
    
    # Normal probability plot
    plt.subplot(1,2,2)
    stats.probplot(df[column], dist="norm", plot=plt)
    plt.title(column)
    
    plt.show()
    
 # Calculate skewness
    skewness = skew(df[column])
    print("Skewness of {}: {}".format(column, skewness))

 # Check for normality
    if abs(skewness) <= 1.96:
        print("The distribution of {} appears to be approximately normal.".format(column))
        normal_features.append(column)
    else:
        print("The distribution of {} does not appear to be approximately normal.".format(column))
        non_normal_features.append(column)
```

Based on the skewness values, it appears that the `residual sugar`, `chlorides`, and `sulphates features` are positively skewed, meaning that the majority of the data points are on the left side of the distribution and there are some extreme values on the right side. To improve normality, we can apply different transformations depending on the data distribution:

- 1. For `residual sugar` and `chlorides`, we can try a l`ogarithmic transformation` or a `square root transformation`, which can compress the extreme values and shift the distribution towards normality.
- 2. `For sulphates`, we can try a `Box-Cox transformation`, which can handle positive skewness better than logarithmic or square root transformations.

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv("winequality-red.csv")

# Separate the features and target variable
X = df.drop("quality", axis=1)
y = df["quality"]

# Standardize the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

## Now that we have preprocessed the data, we can perform PCA to reduce the number of features.

```python
import numpy as np 

# Perform PCA with all components
pca = PCA()
pca.fit(X_std)

# Calculate the cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Find the minimum number of components to explain 90% of the variance
n_components = np.argmax(cumulative_variance_ratio >= 0.9) + 1

print("Minimum number of principal components required to explain 90% of variance:", n_components)
```
